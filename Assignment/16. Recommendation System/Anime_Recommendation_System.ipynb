# Anime Recommendation System using Cosine Similarity

## Objective
To recommend similar anime using cosine similarity based on features like genre and rating.

```python
import pandas as pd
import numpy as np

from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv("anime.csv")
df.head()
```

```python
df.columns = ['anime_id','name','genre','type','episodes','rating','members']
```

```python
# Convert to numeric
df['episodes'] = pd.to_numeric(df['episodes'], errors='coerce')
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Drop missing
df = df.dropna(subset=['genre','rating','episodes'])

# Clean genre text
df['genre'] = df['genre'].str.replace(" ", "")
```

```python
genres = df['genre'].str.get_dummies(sep=',')
```

```python
features = pd.concat([genres, df[['rating','episodes']]], axis=1)
```

```python
scaler = MinMaxScaler()
features[['rating','episodes']] = scaler.fit_transform(features[['rating','episodes']])
```

```python
cosine_sim = cosine_similarity(features)
```

```python
indices = pd.Series(df.index, index=df['name']).drop_duplicates()

def recommend(anime_name, top_n=5):
    if anime_name not in indices:
        return "Anime not found"
    
    idx = indices[anime_name]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    
    sim_scores = sim_scores[1:top_n+1]
    anime_indices = [i[0] for i in sim_scores]
    
    return df['name'].iloc[anime_indices]
```

```python
recommend("Steins;Gate")
```

## Recommendation System Analysis

The recommendation system uses cosine similarity to find similar anime based on genre, rating, and number of episodes.

- Genre is the most influential feature.
- Rating and episode count refine similarity.
- The system recommends anime with similar themes.

Limitations:
- No user personalization
- Popularity bias may exist
