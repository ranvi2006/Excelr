# Estimation, Confidence Intervals and Hypothesis Testing

## Part 1: Confidence Intervals

```python

import numpy as np

data = [1.13,1.55,1.43,0.92,1.25,1.36,1.32,
        0.85,1.07,1.48,1.20,1.33,1.18,1.22,1.29]

mean = np.mean(data)
std = np.std(data, ddof=1)
n = len(data)

mean, std, n
```

### (a) 99% CI using t-distribution

```python

t_value = 2.977
margin = t_value * (std / np.sqrt(n))
lower = mean - margin
upper = mean + margin

lower, upper
```

### (b) 99% CI using z-distribution

```python

sigma = 0.2
z_value = 2.576

margin_z = z_value * (sigma / np.sqrt(n))
lower_z = mean - margin_z
upper_z = mean + margin_z

lower_z, upper_z
```

## Part 2: Hypothesis Testing


### Hypotheses

H0: μ = 4000 (no increase in cost)  
H1: μ > 4000 (cost is higher)


```python

# Given values
sample_mean = 3050
mu = 1000 + 5*600   # theoretical mean
sigma = 5 * 25
n = 25

# Z-test
z = (sample_mean - mu) / (sigma / (n**0.5))
z
```


### Decision

Critical value at 5% = 1.645

If z > 1.645 → reject H0



### Conclusion

Since calculated z is negative and less than critical value,
we fail to reject null hypothesis.

There is no strong evidence that cost is higher than model.
